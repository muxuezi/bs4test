```python
from urllib import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'http://eaie.bjtu.edu.cn/list.php?cid=22&page=1'
eepage = urlopen(url).read()
```

```python
print eepage
```

```python
soup = BeautifulSoup(eepage)
```

```python
tableall = soup.find('div',class_='con')
```

```python
trall = tableall.findAll('tr')
```

```python
len(trall)
```

```python
for idx, tr in enumerate(trall[1:]):
    print idx+1, ';'.join(tr.text.split()), 'http://eaie.bjtu.edu.cn'+tr.find('a').get('href')
```
